```python
s1, s2 = df[(df.Species == "Camallanus cotti") & (df.Gene == "COX3")].Sequence.values
print(s1)
print(s2)
```

```python
s1 == s2
```

```python
n = len(s1)

matches = mismatches = 0
for i in range(n):
    if s1[i] == s2[i]:
        matches += 1
    else:
        mismatches += 1

print(f"matches: {matches}, mismatches: {mismatches}, n: {n}")
```
